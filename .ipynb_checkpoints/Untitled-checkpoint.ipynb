```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import grey_closing, grey_opening
from matplotlib.patches import Rectangle
from skimage.io import imread_collection
from skimage.filters import threshold_otsu
```

```python
def make_sequence(directory, color, label):
    path = 'train/' + directory + '/'
    names = [path + name for name in sorted(os.listdir(path))]
    data = imread_collection(names).concatenate()
    data = data[:, :, :, color].astype('float64')
    data -= np.mean(data)
    data /= np.std(data)
    data[data < 0] = 0
    data = np.gradient(data, axis=0)
    data = np.sum(data, axis=(1, 2)).ravel()
    return data

def draw_images(directory, label):
    plt.style.use('classic')
    path1 = 'train/' + directory + '/'
    path2 = '.' + directory + '.png'
    images = imread_collection([
        path1 + str(label).zfill(3) + path2,
        path1 + str(label + 1).zfill(3) + path2,
        path1 + str(label + 2).zfill(3) + path2
    ]).concatenate().astype('float64')
    images[images < images.mean()] = 0
    [(plt.imshow(image.astype('uint8')), plt.show()) for image in images]
    
def draw_sequence(data, label, title):
    plt.style.use('ggplot')
    plt.plot(data)
    plt.title(title)
    patch = Rectangle((label - 6, data.min()), 12, 
        data.max() - data.min(), fill=False, color='black')
    plt.axes().add_patch(patch)
    plt.show()
```

```python
simple = '030.015'; simple_label = 86
medium = '036.050'; medium_label = 74
hard = '002.084'; hard_label = 116
```

```python
draw_images(simple, simple_label)
```

```python
red = make_sequence(simple, 0, simple_label)
draw_sequence(red, simple_label, 'red') 
green = make_sequence(simple, 1, simple_label)
draw_sequence(green, simple_label, 'green') 
blue = make_sequence(simple, 2, simple_label)
draw_sequence(blue, simple_label, 'blue') 
```

```python
draw_images(medium, medium_label)
```

```python
red = make_sequence(medium, 0, medium_label)
draw_sequence(red, medium_label, 'red') 
green = make_sequence(medium, 1, medium_label)
draw_sequence(green, medium_label, 'green') 
blue = make_sequence(medium, 2, medium_label)
draw_sequence(blue, medium_label, 'blue') 
```

```python
draw_images(hard, hard_label)
```

```python
red = make_sequence(hard, 0, hard_label)
draw_sequence(red, hard_label, 'red') 
green = make_sequence(hard, 1, hard_label)
draw_sequence(green, hard_label, 'green') 
blue = make_sequence(hard, 2, hard_label)
draw_sequence(blue, hard_label, 'blue') 
```
